# 16. Segmentation - Meanshift

Segment some color images using the meanshift algorithm â€“ cv.pyrMeanShiftFiltering(). Try different window
radius (spatial and color), number of pyramid levels and termination criteria.

```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt

DATA_DIR = './data'
FILENAME = 'streetLines_01.jpg'

img = cv2.imread(os.path.join(DATA_DIR, FILENAME))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
for spatialRadius in range(10,50,10):
    segmentedImg = cv2.pyrMeanShiftFiltering(img, spatialRadius, 10)
    plt.imshow(segmentedImg)
    plt.title('spatialRadius = {0}'.format(spatialRadius))
    plt.axis('off')
    plt.show()  
```

```python
for colorRadius in range(10,50,10):
    segmentedRoadImg = cv2.pyrMeanShiftFiltering(img, 20, colorRadius)
    plt.imshow(segmentedRoadImg)
    plt.title('colorRadius = {0}'.format(colorRadius))
    plt.axis('off')
    plt.show()
```

```python
for maxPyrLevel in range(2,12,2):
    segmentedRoadImg = cv2.pyrMeanShiftFiltering(img, 40, 20, maxPyrLevel)
    plt.imshow(segmentedRoadImg)
    plt.title('maxPyrLevel = {0}'.format(maxPyrLevel))
    plt.axis('off')
    plt.show()
```
