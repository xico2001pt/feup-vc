# 19. Morphological operations

```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt

DATA_DIR = './data'
FILENAME = 'roundObjects_03.JPG'

img = cv2.imread(os.path.join(DATA_DIR, FILENAME))

# Invert the image
img = 255 - img

plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
kernel = np.ones((5,5), np.uint8)

# Closing
op = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

plt.imshow(op)
plt.title('After Closing')
plt.axis('off')
plt.show()
```
