# 13. Segmentation - thresholding

Compare the global thresholding â€“ threshold() - and adaptive thresholding - adaptiveThreshold() - for an image
with varying illumination like that in figure 5. Try the several types of thresholding supported by threshold() and
adaptiveThreshold(), namely the Otsu method, in Threshold().

```python
import cv2
import os
import matplotlib.pyplot as plt

DATA_DIR = './data'
FILENAME = 'sudoku.png'

img = cv2.imread(os.path.join(DATA_DIR, FILENAME), cv2.IMREAD_GRAYSCALE)
```

```python
THRESHOLD = 90
MAX_VALUE = 255

# thresholding
ret, thresh1 = cv2.threshold(img, THRESHOLD, MAX_VALUE, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(img, THRESHOLD, MAX_VALUE, cv2.THRESH_BINARY_INV)
ret, thresh3 = cv2.threshold(img, THRESHOLD, MAX_VALUE, cv2.THRESH_TRUNC)
ret, thresh4 = cv2.threshold(img, THRESHOLD, MAX_VALUE, cv2.THRESH_TOZERO)
ret, thresh5 = cv2.threshold(img, THRESHOLD, MAX_VALUE, cv2.THRESH_TOZERO_INV)

# adaptive thresholding
thresh6 = cv2.adaptiveThreshold(img, MAX_VALUE, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
thresh7 = cv2.adaptiveThreshold(img, MAX_VALUE, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

# display images with matplotlib
titles = ['Original Image', 'BINARY', 'BINARY_INV', 'TRUNC', 'TOZERO', 'TOZERO_INV', 'ADAPTIVE_THRESH_MEAN_C', 'ADAPTIVE_THRESH_GAUSSIAN_C']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5, thresh6, thresh7]

for i in range(8):
    plt.subplot(2, 4, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
plt.show()
```
